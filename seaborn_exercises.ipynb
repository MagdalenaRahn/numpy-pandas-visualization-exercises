#### 4 / Use the chipotle dataset from the previous exercise. 
#### Create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
import os

def get_db_url(host, username, password, db):
    url = f'mysql+pymysql://{username}:{password}@{host}/{db}'
    return url
 
url = get_db_url(host, username, password, 'chipotle')    
```

```python
query = 'SELECT * FROM orders'
```

```python
chipo = pd.read_sql(query, url)
```

```python
chipo.head()
```

```python
chipo['price_in_float'] = chipo['item_price'].str.replace('$','').astype('float')
chipo
```

```python
chipo.info()
```

```python
four_most_popular_items = chipo.groupby('item_name')['item_name'].count().nlargest(4)
four_most_popular_items
```

```python
revenue_makers = chipo.groupby('item_name')['price_in_float'].sum().nlargest(4)
revenue_makers
```

```python
sns.lineplot(data = chipo, x = 'item_name', y = 'price_in_float')
plt.xticks(rotation = 90)
```

```python
sns.barplot(x = four_most_popular_items, y = revenue_makers)
```
