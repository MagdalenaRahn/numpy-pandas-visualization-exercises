- Built on top of Matplotlib and integrates Pandas features.
- Built-in datasets : sns.get_dataset_names()... and fill in the blank () with various plot-arguments.
    - 'tips', 'anscombe', 'titanic'...  datasets.  
- Load database :   
    In [ ]:   
      dataset = sns.load_dataset('dataset_name')    

- Various types of visualisation :   
        - boxplots,   
        - scatterplots (kwargs / keyword arguments),  
        - line,  
        - histograms (class on x horizonal, numerical on y vertical axis) (histograms are good for finding
            distribution of data ; kde = True,  
        - heatmaps : sns.heatmap(data, annot = True, cmap = plt.cm.Greens) ; colour darkness on heatmap
            indicates density / correlation of data,  
        - pairplots : all of the quantitative data against all of the quantitative data (no numerical 
            features).  
- Kwargs :
        - Col = 'word'.   sns.relplot(x = 'total_bill', y = 'sth_else', Col = 'variable')   
        - Hue = 'word' ; the legend based on word .  
        - Style = 'word' ; red and black , x and o...
        - KDE = 'True' (includes line) or 'False' (no line). Kernal density estimator ; it's an approximate 
            distribution line.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
# importing 'tips' dataset

tips = sns.load_dataset('tips')

#tips.describe()
```

```python
tips.head()

# total_bill, tip, sex, smoker, day, time, size
```

```python
tips.info()
```

### Scatter plot of tips by total bill, from the tips dataset.

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, s = 100, color = 'b', marker = '*')

plt.title('Tips By Total Bill')
plt.xlabel('Total Bill In $')
plt.ylabel('Tipped Amount in $')

plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', col = 'time', data = tips)

# col = column name based on dataset
```

## Hue

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'size', data = tips)

# size = party size
# hue bases the colour of the datapoint on the bin in which the dp falls
```

```python
sns.relplot(x = 'total_bill', y = 'tip', col = 'time', hue = 'size', style = 'smoker', data = tips)


# adding columns of time, smokers
```

## Histograms / Histplots

```python


sns.histplot(x = 'total_bill', y = 'tip', data = tips, kde = True)

# Histoplot counts amount of instances in one thing
```

```python
sns.histplot(tips.total_bill, kde = True)
```

```python
sns.histplot(tips.tip, kde = True)
```

## Box Plots
Show outliers

```python
sns.boxplot(data = tips, y = 'tip', x = 'size')

# party of 5 has no mean...
```

```python
sns.boxplot(data = tips, y = 'tip', x = 'size', hue = 'sex')
```

## Pair Plots

Comparing

```python
sns.pairplot(data = tips)
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip')
```
