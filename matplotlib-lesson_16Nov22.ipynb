## Line charts


Plt.plot - accepts x values first, y values second.  
Limits - plt.xlim, plt.ylim  
Color - c argument  
Alpha - transparency  
Multiple plots  

```python
# make an array called x

x = np.array(range(1,11))

x
```

```python
# y values, to allow for x and y axes

y = x ** 2

y
```

```python
# let's make a chart

plt.plot(y, c = 'pink')

# we didn't set any x-values. Lists in df are 0-indexed by default. Matplotlib also is 0-index default
```

```python
# just for fun

plt.plot(y, x), plt.plot(x, y)
```

```python
 plt.plot(x, y)
```

```python
# limit the scope of this window on the x axis, from 5 to 10

plt.plot(x,y)
plt.xlim(5,11)
```

```python
# limit the scope of this window on the x axis and y axis

plt.plot(x,y, c = 'magenta')
plt.xlim(5,10)
plt.ylim(60, 100)
```

```python
z = x ** 2 - 3
```

```python
plt.plot(x,y, c = 'darkcyan', alpha = .5)
plt.plot(x,z, c = 'deeppink', alpha = .3)
plt.xlim(4,8)
```

## Scatter charts

Size - s argument

```python
plt.scatter(x,y, c = 'orange', s = 190)
plt.scatter(x,z, c = 'aqua', s = 10)

# adding a title and axes names
plt.title('Upside-down Oranges')
plt.xlabel('Unknown X Values')
plt.ylabel('Unknown Y Values')
# default is s = 40
```

```python
plt.scatter(x,y, c = 'orange', s = 190)
plt.plot(x,z, c = 'aqua')
```

## Custom Line Types

Line style - ls argument (accepts ":" and "- -")

```python
plt.plot(x,y, ls = "--")
plt.plot(x,y, ls = ":")
plt.plot(z,y, ls = ":")
```

## Rotating text and annotations

Rotation - rotation argument

```python
plt.plot(x,y, c = 'darkcyan', alpha = .5, ls = ":")
plt.plot(x,z, c = 'deeppink', alpha = .3, ls = "--")
plt.xticks(rotation = 50)
```

## Adding text

plt.text - x and y coordinates, text, fontsize, color

plt.annotate - text, pointer coordinates, text coordinates, arrow properties

```python
plt.plot(x,y, c = 'darkcyan', alpha = .5, ls = ":")
plt.plot(x,z, c = 'deeppink', alpha = .3, ls = "--")
plt.text(2,19,'essai', fontsize = 30, c = 'orange', alpha = .6)
# this makes a word appear at the point of x-y coordinate location.
```

```python
plt.plot(x,y, c = 'darkcyan', alpha = .5, ls = ":")
plt.plot(x,z, c = 'deeppink', alpha = .3, ls = "--")
plt.annotate(text = ' Regarde !', xy = (2,10), xytext = (3,8), arrowprops = {'width': 3})
plt.title('Encore un essai')
# plt.annotate is a way to add more complext text to the chart

# arrowprops dictionary
```

## Figure size and legends

plt.figure - figsize argument to determine size of plot  

plt.legend - loc argument for location

```python
plt.figure(figsize = (4, 4))
# changes the size of graph

plt.plot(x,y, c = 'darkcyan', alpha = .5, ls = ":")
plt.plot(x,z, c = 'deeppink', alpha = .3, ls = "--")
```

```python
plt.figure(figsize = (4, 4), dpi = 40)
# changes the dpi

plt.plot(x,y, c = 'darkcyan', alpha = .5, ls = ":")
plt.plot(x,z, c = 'deeppink', alpha = .3, ls = "--")
```

```python
plt.figure(figsize = (4,4))
# changes the size of graph

plt.plot(x,y, c = 'darkcyan', alpha = .5, ls = ":")
plt.plot(x,z, c = 'deeppink', alpha = .3, ls = "--")

# plt.legend([1,2])

plt.legend(['dot', 'dash'])
```

```python
plt.figure(figsize=(12, 8))
for i in range(1, 5):
    n = 10 ** i
    x1 = [randint(1, 6) for _ in range(n)]
    x2 = [randint(1, 6) for _ in range(n)]

    total = [x + y for x, y in zip(x1, x2)]

    plt.subplot(2, 2, i)
    plt.hist(total, bins=range(2, 14), align='left', color='white', edgecolor='black')
    plt.title(f'n = {n}')

plt.suptitle('Sum of 2 dice rolls')
plt.show()

```
