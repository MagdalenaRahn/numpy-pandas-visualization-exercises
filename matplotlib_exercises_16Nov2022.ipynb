```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from random import randint
import math as m

%matplotlib inline
```

```python
# 1 / Use matplotlib to plot the following equation: y = x2 − x + 2

# Write the code that generates the x and y points.

# Add an anotation for the point 0, 0, the origin.

x = np.array(range(0,15))
x
```

```python
# 1
plt.plot(x,y, c = 'g')

plt.annotate(text = ' L\'Origine est à zéro', xy = (0,0), xytext = (6,10), arrowprops = {'width': 3})
```

```python
# 2 / Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
''' 
y=√x
y=x3
y=2x
y=1/(x+1)

You can use functions from the math module to help implement some of the equations above.
'''
```

```python
# 2 
x = np.array(range(0,15))
y_sq = np.sqrt(x)
y_sq


# y = x ** (0.5)
```

```python
# 2

plt.plot(x, y_sq)
plt.title('Y Is The Square Root Of X')

plt.xticks(range(0,16))
plt.yticks(range(0, 5))
```

```python
# 2 / y=x3
y3 = x * 3
y3
```

```python
plt.plot(x,y3)

plt.title('X Times 3 Is Y')

plt.xticks(range(0,16))
plt.yticks(range(0, 45,4))
```

```python
# 2 / y=2x

y2 = x * 2
y2
```

```python
plt.plot(x,y2, c = 'y')

plt.title('X Times 2 Is Y')

plt.xticks(range(0,16))
plt.yticks(range(0, 32,4))

plt.figure(figsize = (4, 4))
```

```python
# 2 / y=1/(x+1)

y1 = 1 / (x + 1)
y1
```

```python
# 2 / 

plt.plot(x,y1, c = 'turquoise')

plt.title('Y Is 1 Divided by X Plus 1')

plt.xticks(range(0,16))
plt.yticks(range(0, 2))

```

```python
# 3 / Combine the figures you created in the last step into one large figure with 4 subplots.

plt.subplot(1,4,1)
plt.plot(x, y_sq, c = 'lime')

plt.subplot(1,4,2)
plt.plot(x, y3, c = 'magenta')

plt.subplot(1,4,3)
plt.plot(x, y2, c = 'y')

plt.subplot(1,4,4)
plt.plot(x, y1, c = 'c')


plt.subplots_adjust(left = 0.01,
                    bottom = 0.1,
                    right = 1,
                    top = 0.9,
                    wspace = 0.5,
                    hspace = 0.4)

# plt.subplot - number of rows, number of columns, selected figure
```

```python
# 4 / 
# Combine the figures you created in the last step into one figure where each of the 4 equations 
# has a different color for the points. 
# Be sure to include a legend and an appropriate title for the figure.

(
        plt.plot(x, y_sq, c = 'lime'), 
        plt.plot(x, y3, c = 'magenta'), 
        plt.plot(x, y2, c = 'y'), 
        plt.plot(x, y1, c = 'c')
)
plt.title('X And Y Have Four Conversations')

plt.legend(['y = √x','y = x3', 'y = 2x','y = 1 / (x + 1)'])
```
